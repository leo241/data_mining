# 一、贝叶斯分类器

## 1.极大似然估计

先验极大似然估计的概率：

$P(Y = c_k) = \frac{\Sigma_{i=1}^N I(y_i = c_k)}{N}$

在本题中有，

$P(Y = 1) = \frac{5}{15} = \frac{1}{3}$ \
$P(Y = -1) = \frac{10}{15} = \frac{2}{3}$

条件概率极大似然估计：

$P(X^{(j)} = a_{jl}|Y = c_k) = \frac{\Sigma_i I(x_i^{(j)} = a_{jl}, y_i = c_k)}{\Sigma_i I(y_i = c_k)}$

在本题中有，

$P(X^{(1)} = 1|Y = 1) = \frac{1}{5}$ \
$P(X^{(1)} = 2|Y = 1) = \frac{3}{5}$ \
$P(X^{(1)} = 3|Y = 1) = \frac{1}{5}$ \
$P(X^{(1)} = 1|Y = -1) = \frac{2}{10}=\frac{1}{5}$ \
$P(X^{(1)} = 2|Y = -1) = \frac{2}{10}=\frac{1}{5}$ \
$P(X^{(1)} = 3|Y = -1) = \frac{6}{10}=\frac{3}{5}$

$P(X^{(2)} = S|Y = 1) = \frac{1}{5}$ \
$P(X^{(2)} = M|Y = 1) = \frac{2}{5}$ \
$P(X^{(2)} = L|Y = 1) = \frac{2}{5}$ \
$P(X^{(2)} = S|Y = -1) = \frac{3}{10}$ \
$P(X^{(2)} = M|Y = -1) = \frac{4}{10}=\frac{2}{5}$ \
$P(X^{(2)} = L|Y = -1) = \frac{3}{10}$

对于题目中给定$x = (2,M)^T$

$P(Y = 1|x)的分子 = P(Y = 1)P(X^{(1)} = 2|Y = 1)P(X^{(2)} = M|Y = 1) = \frac{1}{3} \times \frac{3}{5} \times \frac{2}{5} = \frac{2}{25}$ \
$P(Y = -1|x)的分子 = P(Y = -1)P(X^{(1)} = 2|Y = -1)P(X^{(2)} = M|Y = -1) = \frac{2}{3} \times \frac{1}{5} \times \frac{2}{5} = \frac{4}{75}$

$P(Y = 1|x) > P(Y = -1|x)$

故样本$x = (2,M)^T$的分类是$Y = 1$

## 2.贝叶斯估计

记$K$为Y的可能取值个数，$S_j$为$x^{(j)}$的可能取值个数，则有

先验概率：

$P_\lambda(Y = c_k) = \frac{\Sigma_{i=1}^N I(y_i = c_k) + \lambda}{N + K\lambda}$

条件概率估计：

$P_\lambda(X^{(j)} = a_{jl}|Y = c_k) = \frac{\Sigma_i I(x_i^{(j)} = a_{jl}, y_i = c_k) + \lambda}{\Sigma_i I(y_i = c_k)+ S_j \lambda}$

本题中取$\lambda = 1$，则有：

$K = 2, S_1 = 3, S_2 = 3$

$P_\lambda(Y = 1) = \frac{5 + 1}{15 + 2} = \frac{6}{17}$ \
$P_\lambda(Y = -1) = \frac{10 + 1}{15 + 2} = \frac{11}{17}$

$P_\lambda(X^{(1)} = 1|Y = 1) = \frac{1+1}{5+3} = \frac{1}{4}$ \
$P_\lambda(X^{(1)} = 2|Y = 1) = \frac{3+1}{5+3}= \frac{1}{2}$ \
$P_\lambda(X^{(1)} = 3|Y = 1) = \frac{1+1}{5+3}= \frac{1}{4}$ \
$P_\lambda(X^{(1)} = 1|Y = -1) = \frac{2+1}{10+3}=\frac{3}{13}$ \
$P_\lambda(X^{(1)} = 2|Y = -1) = \frac{2+1}{10+3}=\frac{3}{13}$ \
$P_\lambda(X^{(1)} = 3|Y = -1) = \frac{6+1}{10+3}=\frac{7}{13}$

$P_\lambda(X^{(2)} = S|Y = 1) = \frac{1+1}{5+3}= \frac{1}{4}$ \
$P_\lambda(X^{(2)} = M|Y = 1) = \frac{2+1}{5+3}= \frac{3}{8}$ \
$P_\lambda(X^{(2)} = L|Y = 1) = \frac{2+1}{5+3}= \frac{3}{8}$ \
$P_\lambda(X^{(2)} = S|Y = -1) = \frac{3+1}{10+3}= \frac{4}{13}$ \
$P_\lambda(X^{(2)} = M|Y = -1) = \frac{4+1}{10+3}=\frac{5}{13}$ \
$P_\lambda(X^{(2)} = L|Y = -1) = \frac{3+1}{10+3}= \frac{4}{13}$

对于题目中给定$x = (2,M)^T$

$P_\lambda(Y = 1|x)的分子 = P_\lambda(Y = 1)P_\lambda(X^{(1)} = 2|Y = 1)P_\lambda(X^{(2)} = M|Y = 1) = \frac{6}{17} \times \frac{1}{2} \times \frac{3}{8} = \frac{9}{136} = 0.066$\
$P_\lambda(Y = -1|x)的分子 = P_\lambda(Y = -1)P_\lambda(X^{(1)} = 2|Y = -1)P_\lambda(X^{(2)} = M|Y = -1) = \frac{11}{17} \times \frac{3}{13} \times \frac{5}{13} = \frac{165}{2873} = 0.057$

$P_\lambda(Y = 1|x) > P_\lambda(Y = -1|x)$

故样本$x = (2,M)^T$的分类是$Y = 1$

## 3.分词+词云

导入额外自定义词典、对于生鲜进行分词、提取top100高频词汇制作词云：

```python
import jieba
from collections import Counter
from wordcloud import WordCloud
names = list(catalog[catalog['first'] == '生鲜']['name'])
names = list(map(str, names))
jieba.load_userdict('data/userdict.dat')
dic = Counter(jieba.lcut(''.join(names))).most_common(100)
top100 = [item[0] for item in dic]
wc = WordCloud(font_path='msyh.ttc').generate(' '.join(top100)) # 需要在目录下有中文字体文件'msyh.ttc'
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

## 4.预处理

去掉纯字母

```python
import re
df = catalog[catalog['first'] == '食品饮料、保健食品']
def tokenize(x):
    x = str(x)
    stop_word_pattern = "\(|\)|[a-zA-Z0-9]"
    x = re.sub(stop_word_pattern, '', x)
    return jieba.lcut(x)
df['tokenize'] = df['name'].apply(lambda x: tokenize(x))
```

对词频进行截断

```python
from tqdm import tqdm
def get_vocabulary(df, low = 10, high = 0.75, top = 1000):
    all_words = []
    high = high * len(df)
    for item in df['tokenize']:
        all_words += item
    s = list(set(all_words))
    dic2 = dict()
    for i in s:
        dic2[i] = 0
    for item in df['tokenize']:
        for word in list(set(item)):
            dic2[word] += 1
    s2 = []
    for i in s:
        if dic2[i] >= low and dic2[i] <= high:
            s2.append(i)
    new_words = []
    for word in tqdm(all_words):
        if word in s2:
            new_words.append(word)
    dic = Counter(new_words)
    cb = dic.most_common(top)
#     voc = [i[0] for i in cb]
    return cb
cb = get_vocabulary(df)
```

```python
import numpy as np
voc = [item[0] for item in cb]
x = []
for item in tqdm(df['tokenize']):
    x.append([1 if word in item else 0 for word in voc])
```

```python
x = np.asarray(x)
```

## 5.模型训练与评估

```python
from sklearn.model_selection import train_test_split
y = df['second']
dic = list(y.value_counts().keys())
dic2 = dict()
for i, v in enumerate(dic):
    dic2[v] = i
y = y.apply(lambda x: dic2[x])
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB().fit(xtrain,ytrain)
```

```python
ypred = gnb.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(ytest, ypred))
print(accuracy_score(ytest, ypred))
```

从准确率和困惑矩阵看出当前的效果不是很好
